```python
import numpy as np
import scipy
import read_mist_models
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
iso00 = []

for i in range(100):
    iso = read_mist_models.ISOCMD('metal0_extinction0.iso.cmd')
    age_ind = iso.age_index(5.0+i*.05)
    I00 = iso.isocmds[age_ind]['ACS_WFC_F814W']
    R00 = iso.isocmds[age_ind]['ACS_WFC_F606W']
    iso00.append((R00-I00, R00))
    plt.plot(R00-I00, R00, alpha = .5, color = matplotlib.cm.inferno(i/100))
    
iso00 = np.asarray(iso00)
```

```python
for i in range(20):
    iso = read_mist_models.ISOCMD('metal0_extinction0.iso.cmd')
    age_ind = iso.age_index(5.0+i*.25)
    Teff = iso.isocmds[age_ind]['log_Teff']
    L = iso.isocmds[age_ind]['log_L']
    plt.plot(-Teff,L) 
```

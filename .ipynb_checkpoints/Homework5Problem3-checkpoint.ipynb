```python
import numpy as np
import matplotlib.pyplot as plt
import fsps
%matplotlib inline

alldata = np.genfromtxt('ps5.data', usecols = range(15))
np.shape(alldata)
gdata = np.empty([463, 10])
gdata[:,0] = alldata[:,1]*3600+alldata[:,2]*60+alldata[:,3]
gdata[:,1] = (np.abs(alldata[:,4])*3600+alldata[:,5]*60+alldata[:,6])*np.sign(alldata[:,4])
gdata[:,2:] = alldata[:,7:]
gdata
```

```python
metallicity = 6.27 - .11*(gdata[:,3]-5*(np.log10(gdata[:,2]*1e5))) # converting from apparent to absolute magnitude
plt.figure(figsize=(10,10))
plt.hist(metallicity, bins = 20)
plt.xlabel('12 + log(O/H)', fontsize = 20)
plt.ylabel('Number', fontsize = 20)
plt.title('Metallicity Distribution in GALEX Sample', fontsize = 20)
```

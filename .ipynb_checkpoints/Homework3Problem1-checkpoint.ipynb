```python
iso00 = []

for i in range(100):
    iso = read_mist_models.ISOCMD('metal0_extinction0.iso.cmd')
    age_ind = iso.age_index(5.0+i*.05)
    I00 = iso.isocmds[age_ind]['ACS_WFC_F814W']
    R00 = iso.isocmds[age_ind]['ACS_WFC_F606W']
    iso00.append((R00-I00, R00))
    plt.plot(R00-I00, R00, alpha = .5, color = matplotlib.cm.inferno(i/100))
    
iso00 = np.asarray(iso00)
```

And I think I will leave my assessment here. The PARSEC fits end up age = 11 billion years--the upper limit age in the model-- and a metallicity of -2.1 Z_sun, along with an extinction of apparently 1 magnitude. None of these fits get the turnoff quite right; they are too sharp. The problem was also present in the MIST models, but is more pronounced here. Additionally, these models don't try to fit the horizontal branch, while they overfit the red giant branch. With the MIST models, I used the horizontal branch to fit the model by eye, but in the PARSEC fits, all I have is the MS turnoff. One reason for the major differences in age and extinction is likely because of the upper limit placed on the age by the model. A higher extinction is the only way I could move the trend line on top of the actual data.
